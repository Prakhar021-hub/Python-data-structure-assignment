Q9.ezplain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
Ans:The normal distribution (also known as the Gaussian distribution) is a continuous probability distribution that is symmetrical and bell-shaped. It is one of the most important and widely used distributions in statistics because many natural phenomena follow this distribution, and many statistical methods assume data is normally distributed.

Properties of the Normal Distribution:
1.Symmetry:
The normal distribution is perfectly symmetrical about its mean, which means the left side of the curve is a mirror image of the right side.
Mean = Median = Mode: In a normal distribution, these three measures of central tendency are equal and located at the center of the distribution.

2.Bell-shaped Curve:
The distribution forms a bell-shaped curve, with most of the data points clustering around the mean and fewer data points appearing as you move further from the mean.

3.Tails Approach Zero:
The tails of the normal distribution extend indefinitely in both directions but get closer and closer to the horizontal axis without ever touching it. This indicates that extremely large or small values are possible but less likely.

4.Defined by Mean and Standard Deviation:
The shape of the normal distribution is entirely defined by its mean (which determines the center) and its standard deviation (which determines the spread or width of the distribution).
A larger standard deviation results in a wider and flatter curve, while a smaller standard deviation results in a narrower and taller curve.

5.Area Under the Curve:
The total area under the normal distribution curve is equal to 1, representing 100% of the probability. This property allows the calculation of probabilities for different intervals within the distribution.


The Empirical Rule (68-95-99.7 Rule):
The Empirical Rule, or 68-95-99.7 rule, is a shorthand way to describe the distribution of data in a normal distribution based on standard deviations from the mean.
It states that the 68% of the data lies in 1st standard deviation, 95% of the data lies in 2nd standard deviation, and 99.7% of the data lies in the 3rd standard deviation.
